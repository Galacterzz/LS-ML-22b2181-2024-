# Neccessary imports

```python
from ultralytics import YOLO
```

# Loading model

```python
# Path to the dataset YAML file
data_path = r'C:\Users\ASUS\OneDrive\Desktop\desktop folders\TECH\LS\ML\code\week 2\YOLOv8 Assign\car-20240712T082202Z-001\car\dataset.yaml'
```

```python
# Load YOLOv8 model
model = YOLO('yolov8n.pt')
```

# Training YOLOv8 Model

```python
# Train the model on your custom dataset
model.train(data = data_path, epochs = 25)
```

```python
# Validate the model
model.val(data=data_path)
```

# Testing Model

```python
# Evaluate the model on the test set
model.val(data=data_path, split='test')
```
